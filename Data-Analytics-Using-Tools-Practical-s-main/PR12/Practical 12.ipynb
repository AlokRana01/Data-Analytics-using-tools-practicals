```python
# ==============================
# PR 12: DATA CHARACTERIZATION
# ==============================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# Step 1: Load Dataset
# ==============================
df = pd.read_csv('cleaned_data.csv')

print("\n🔹 First 5 Rows:")
print(df.head())

# ==============================
# Step 2: Data Types & Summary
# ==============================
print("\n🔹 Data Types:")
print(df.dtypes)

print("\n🔹 Summary Statistics:")
print(df.describe(include='all'))

# ==============================
# Step 3: Missing Values
# ==============================
print("\n🔹 Missing Values Count:")
print(df.isnull().sum())

print("\n🔹 Missing Percentage:")
missing_percentage = df.isnull().mean() * 100
print(missing_percentage)

# ==============================
# Step 4: Unique Values
# ==============================
print("\n🔹 Unique Values in Categorical Columns:")
for col in df.select_dtypes(include='object'):
    print(f"\nColumn: {col}")
    print(df[col].unique())

# ==============================
# Step 5: Combine Data (Demo)
# ==============================
df1 = df.copy()
df2 = df.copy()

df_concat = pd.concat([df1, df2], ignore_index=True)
print("\n🔹 Concatenated Shape:", df_concat.shape)

df_merge = pd.merge(df1, df2, on='EmployeeID', how='inner')
print("🔹 Merged Shape:", df_merge.shape)

# ==============================
# Step 6: Visualize Missing Data (Seaborn)
# ==============================
print("\n🔹 Missing Data Visualization")

plt.figure()
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Heatmap")
plt.show()

# ==============================
# Step 7: Categorical Distribution
# ==============================
print("\n🔹 Categorical Distribution (%):")
for col in df.select_dtypes(include='object'):
    print(f"\n{col}:")
    print(df[col].value_counts(normalize=True) * 100)

# ==============================
# Step 8: Correlation Analysis
# ==============================
print("\n🔹 Correlation Matrix:")

corr = df.corr(numeric_only=True)
print(corr)

plt.figure()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# ==============================
# Final Message
# ==============================
print("\n✅ PR 12 Completed Successfully!")
```
